```python
from pathlib import Path
import re
import numpy
import csv
import json
import os
import datetime


pattern = re.compile(r'.*Divvy_Trips_.*Divvy_Trips_.*\.csv$')
out_file = "trip_summary.csv"

with open(out_file, 'a') as fout:
    wr = csv.writer(fout)
    wr.writerow(["starttime","stoptime","bikeid","tripduration",
        "from_station_id","from_station_name","to_station_id",
        "to_station_name","usertype","gender","birthday"])

    for year_dir in Path("./data").iterdir():
        if year_dir.is_dir():
            for trip_file in year_dir.iterdir():
                # print (trip_file)
                match = re.match(pattern, str(trip_file))
                if match:
                    # print (trip_file)
                    with open(out_file, 'a') as outfile, open(trip_file, 'r') as infile:
                        try: 
                            reader = csv.reader(infile)
                            next(reader, None)
                            writer = csv.writer(outfile)
                            for row in reader:
                                writer.writerow(row[1:])
                        except:
                            print("failed to read {}".format(trip_file.name))
start_year = []
start_month = []
start_day = []
start_hour = []
start_minute = []
end_year = []
end_month = []
end_day = []
end_hour = []
end_minute = []
bikeid = []
tripduration = []
from_station_id = []
from_station_name = []
to_station_id = []
to_station_name = []
usertype = []
gender = []
birthday = []

with open("./trip_summary.csv", 'r') as fp:
    reader = csv.reader(fp)
    next(reader, None)
    for row in reader:
        try:
            t1 = datetime.datetime.strptime(row[0], "%Y-%m-%d %H:%M")
            start_year.append(t1.year)
            start_month.append(t1.month)
            start_day.append(t1.day)
            start_hour.append(t1.hour)
            start_minute.append(t1.minute)
        except ValueError:
            try:
                t1 = datetime.datetime.strptime(row[0], "%m/%d/%Y %H:%M")
                start_year.append(t1.year)
                start_month.append(t1.month)
                start_day.append(t1.day)
                start_hour.append(t1.hour)
                start_minute.append(t1.minute)
            except ValueError:
                try:
                    t1 = datetime.datetime.strptime(row[0], "%m/%d/%Y %H:%M:%S")
                    start_year.append(t1.year)
                    start_month.append(t1.month)
                    start_day.append(t1.day)
                    start_hour.append(t1.hour)
                    start_minute.append(t1.minute)
                except:
                    pass

        try:
            t2 = datetime.datetime.strptime(row[1], "%Y-%m-%d %H:%M")
            end_year.append(t2.year)
            end_month.append(t2.month)
            end_day.append(t2.day)
            end_hour.append(t2.hour)
            end_minute.append(t2.minute)
        except ValueError:
            try:
                t2 = datetime.datetime.strptime(row[1], "%m/%d/%Y %H:%M")
                end_year.append(t2.year)
                end_month.append(t2.month)
                end_day.append(t2.day)
                end_hour.append(t2.hour)
                end_minute.append(t2.minute)
            except ValueError:
                try:
                    t2 = datetime.datetime.strptime(row[1], "%m/%d/%Y %H:%M:%S")
                    end_year.append(t2.year)
                    end_month.append(t2.month)
                    end_day.append(t2.day)
                    end_hour.append(t2.hour)
                    end_minute.append(t2.minute)
                except:
                    pass

        bikeid.append(row[2])
        tripduration.append(row[3])
        from_station_id.append(row[4])
        from_station_name.append(row[5])
        to_station_id.append(row[6])
        to_station_name.append(row[7])
        usertype.append(row[8])
        if row[9] == "":
            gender.append('Unknown')
        else:
            gender.append(row[9])
        if row[10] == "":
            birthday.append('NA')
        else:
            birthday.append(row[10])

print(len(start_year))
print(len(bikeid))

print(gender[:10])
```

```python
import csv
import json

csvfile = open('trip.csv', 'r')
jsonfile = open('trip.json', 'w')

fieldnames = ('trip_id', 'start_year', 'start_month', 'start_day', 'start_hour', 'start_minute',
    'end_year', 'end_month', 'end_day', 'end_hour', 'end_minute', 'bikeid', 'tripduration',
    'from_station_id', 'from_station_name', 'to_station_id', 'to_station_name', 'usertype',
    'gender', 'birthday')
reader = csv.DictReader(csvfile, fieldnames)
for row in reader:
    json.dump(row, jsonfile, indent = 2)
    jsonfile.write('\n')
csvfile.close()
jsonfile.close()
```
